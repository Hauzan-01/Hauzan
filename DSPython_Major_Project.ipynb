```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['_golden2'] = le.fit_transform(df['_golden'])
df['_unit_state2'] = le.fit_transform(df['_unit_state'])
df['name2'] = le.fit_transform(df['name'])

df['gender2'] = df['tweet_count']
for i in range(20050):
    if(df['gender'][i] == 'male'):
        df['gender2'][i] = 0
    elif(df['gender'][i] =='female'):
        df['gender2'][i] = 1
    else:
        df['gender2'][i] = 2
        
for i in range(20050):
    if(df['gender:confidence'][i]==' '):
        df['gender:confidence'][i] = 0
df['gender:confidence2'] = le.fit_transform(df['gender:confidence'])
    
df
```

```python
df['gender:confidence2']
```

```python
X = df[['_golden2','_unit_state2','_trusted_judgments','gender:confidence2','profile_yn:confidence','fav_number','name2','retweet_count','tweet_count']].values
Y = df[['gender2']].values
```

```python
from sklearn.model_selection import train_test_split
#Logistic Regression
```

```python
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X, Y, train_size=0.8)
```

```python
X_train1.shape, X_test1.shape, Y_train1.shape, Y_test1.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogReg = LogisticRegression()
```

```python
LogReg.fit(X_train1, Y_train1)
```

```python
y_pred1 = LogReg.predict(X_test1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred1, Y_test1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(Y_test1,y_pred1))
```

```python
# Knn Classifier
X_train2, X_test2, Y_train2, Y_test2 = train_test_split(X, Y, train_size=0.8)
```

```python
X_train2.shape, X_test2.shape, Y_train2.shape, Y_test2.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
knn
```

```python
knn.fit(X_train2,Y_train2)
```

```python
y_pred2 = knn.predict(X_test2)
```

```python
accuracy_score(y_pred2, Y_test2)
```

```python
print(classification_report(Y_test2,y_pred2))
```

```python
# Random Forest Classifier
X_train3, X_test3, Y_train3, Y_test3 = train_test_split(X, Y, train_size=0.8)
```

```python
X_train3.shape, X_test3.shape, Y_train3.shape, Y_test3.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train3,Y_train3)
```

```python
y_pred3 = rfc.predict(X_test3)
```

```python
accuracy_score(y_pred3, Y_test3)
```

```python
print(classification_report(Y_test3,y_pred3))
```
